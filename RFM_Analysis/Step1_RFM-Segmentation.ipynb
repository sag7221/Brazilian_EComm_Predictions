```python
#check outliers that fall with +/- 5% boundaries
def num_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.05)
    q3 = df_in[col_name].quantile(0.95)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    outliers_df= df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)]
    return print("Number of outliers in {} column: ".format(col_name), len(outliers_df)), print("Indexes: ", outliers_df.index)
```

```python
for i in ["Recency", "Frequency", "Monetary"]:
    print(num_outlier(RFM_df, i))
```

```python
#remove outliers that fall with +/- 5% boundaries, except Frequency columns:
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.05)
    q3 = df_in[col_name].quantile(0.95)
    iqr = q3-q1     
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    index_outliers= df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)].index
    df_in= pd.DataFrame(df_in.drop(index_outliers.to_list(), axis=0, inplace=True))
    print("Outliers in the {} column have been removed".format(col_name))
    return df_in
```

```python
#FEATURE ENGINEERING
# 1. review_score_classification

df.review_score.unique()
```

```python
def cat_review_score(x):
    if x >= 4:
        return 'positive'
    elif 1<=x<=2:
        return 'negative'
    else:
        return 'neutral'
    
df['review_score_classification'] = df.review_score.apply(lambda baris: cat_review_score(baris))

df.head()
```

```python
# 2. delivery_time

df['delivery_time'] = df['order_delivered_customer_date'] - df['order_purchase_timestamp']
df.head()
```

```python
# 3. delivery_time_classification

def cat_delivery_time(x):
    if x >= 7 :
        return 'slow'
    elif 4<=x<=6:
        return 'medium'
    else:
        return 'fast'
    
df['delivery_time_classification'] = df['delivery_time'].dt.days.apply(lambda x: cat_delivery_time(x))
```

```python
df.head()
```

```python
df['delivery_time_classification'].value_counts().plot.pie(autopct='%1.2f%%', figsize=(8, 8))
```

```python
# 4. review_upload_time

df['review_upload_time'] = df['review_answer_timestamp'] - df['review_creation_date']
df.head()
```

```python
# 5. seller_response

def respon_seller(x):
    if x > 3 :
        return 'slow'
    elif x == 0 :
        return 'on time'
    else:
        return 'fast'
    
df['seller_response'] = df['review_upload_time'].dt.days.apply(lambda x: respon_seller(x))
df.head()
```

```python
def cat_payment_value(x):
    if x >= 700.00:
        return 'expensive'
    elif 200.00<=x<=699.99:
        return 'medium'
    else:
        return 'cheap'
    
df['payment_value_classification'] = df.payment_value.apply(lambda x: cat_payment_value(x))
df.head()
```

```python
# 7. debt_customer

def cat_customer_debt(x):
    if x == 1:
        return False
    else:
        return True

df['debt_customer'] = df.payment_installments.apply(lambda x: cat_customer_debt(x)) 
df.head()
```

```python
# 8. cash_customer

def cat_customer_bayar_cash(x):
    if x == ('boleto'):
        return True
    elif x == ('debit_card'):
        return True
    else:
        return False
    
df['cash_customer'] = df.payment_type.apply(lambda x: cat_customer_bayar_cash(x))   
df.head()
```

```python
# 9. product_width_classification

def cat_width(x):
    if x > 3700:
        return 'heavy'
    elif 1700<=x<=3700:
        return 'medium'
    else:
        return 'light'
    
df['product_width_classification'] = df.product_weight_g.apply(lambda x: cat_width(x))
df.head()
```
